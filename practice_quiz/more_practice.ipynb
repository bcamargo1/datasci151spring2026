```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# 1. Create a dummy CSV file for Q1 and Q3
data = {
    'Day': range(1, 21),
    'Temperature': [random.uniform(60, 90) for _ in range(20)],
    'Humidity': [random.uniform(30, 80) for _ in range(20)],
    'WindSpeed': [random.uniform(0, 15) for _ in range(20)]
}
df = pd.DataFrame(data)
df.to_csv('weather_data.csv', index=False)

print("Setup complete! 'weather_data.csv' has been created.")
```

Q1: Read in data and create a Line Plot

Task:

Import pandas and matplotlib.pyplot.

Read the file "weather_data.csv" (created above) into a DataFrame named df_weather.

Create a line plot (not a scatter plot) with:

Day on the x-axis.

Temperature on the y-axis.

Label the x-axis "Day of Month" and the y-axis "Temperature (F)".

Add the title "Daily Temperature Trend".

```python
import pandas as pd
import matplotlib.pyplot as plt

df_weather = pd.read_csv("weather_data.csv")

plt.plot(df_weather['Day'], df_weather['Temperature'])
plt.xlabel('Day of Month')
plt.ylabel('Temperature')
plt.title('Daily Temperature Trend')
plt.show()
```

Task:

Create a variable keyword2 with the value "Invoice".

Write an if/elif/else block:

If keyword1 is in email_subject, print "High Priority".

Elif keyword2 is in email_subject, print "Finance Related".

Try changing email_subject to test different outcomes.
